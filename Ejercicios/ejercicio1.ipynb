```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Ejercicio 1

a) Implementar un Generador Congruencial Lineal (GCL) de módulo 2^32, multiplicador 1013904223, incremento de 1664525 y semilla  igual a la parte entera del promedio de los números de padrón de los integrantes del grupo.

```python
#Defino los parametros iniciales
modulo = 2**32
multiplicador = 1013904223
incremento = 1664525

#Creo una lista donde vamos a poner los resultados
numeros_aleatorios = []

#Genero el primer numero en base al promedio de los padrones (semilla inicial)
u0 = int(np.mean([104350, 100985, 97524, 85881]))

#Lo agrego a la lista
numeros_aleatorios.append(u0)


#Defino el tamanio de la muestra a usar
tamanio_muestra = 100000

#Genero los nuevos numeros aleatorios implementando un GCL
# Un+1 = (multiplicador * Un + incremento) % modulo

for i in range(tamanio_muestra):
    u1 = (multiplicador * u0 + incremento) % modulo
    numeros_aleatorios.append(u1)
    u0 = u1
    
```

```python
numeros_aleatorios
```

b) Modificar el GCL implementado en el punto a) para que devuelva números al azar entre 0 y 1

```python
#Creo el generador entre 0 y 1, diviendo cada numero anterior por la suma de todos.
numeros_aleatorios_entre_0_y_1 = [i/sum(numeros_aleatorios) for i in numeros_aleatorios]
```

c) Realizar los gráficos que considere necesarios para mostrar las distribuciones de números al azar generados en los puntos a) y b)

```python
#Grafico el primero punto, generado por numeros aleatorios
plt.figure(figsize=(11.7, 8.27))

plt.hist(numeros_aleatorios)
plt.title("Grafico Ejercicio A", fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(11.7, 8.27))

ax = sns.distplot(numeros_aleatorios, kde=False)
ax.set_title("Grafico Ejercicio A",fontsize=30)
```

```python
#Grafico el primero punto, generado por numeros aleatorios
plt.figure(figsize=(11.7, 8.27))

plt.hist(numeros_aleatorios_entre_0_y_1)
plt.title("Grafico Ejercicio B", fontsize=30)
plt.show()
```

```python
plt.figure(figsize=(11.7, 8.27))

ax = sns.distplot(numeros_aleatorios_entre_0_y_1, kde=False)
ax.set_title("Grafico Ejercicio B",fontsize=30)
```
